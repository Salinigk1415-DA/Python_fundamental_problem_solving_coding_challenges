## **College Event Management System**
Create a list called participants and perform add, insert, remove, and display operations.

Create a tuple called events and display the first event, last event, and total number of events.

Create a list called departments, add duplicate department names, convert the list into a set to remove duplicates, and perform add and remove operations.

Create a list called winners using tuples and convert the winners list into a frozenset.


```python
# College Event Management System

# 1. Participants List Operations
participants = ["Arun", "Priya", "Kavin"]

print("Initial Participants:", participants)

# Add participant
participants.append("Meena")
print("After Adding:", participants)

# Insert participant
participants.insert(1, "Rahul")
print("After Inserting:", participants)

# Remove participant
participants.remove("Kavin")
print("After Removing:", participants)

# Display participants
print("\nFinal Participants List:")
for p in participants:
    print(p)

# 2. Events Tuple Operations
events = ("Coding Contest", "Paper Presentation", "Quiz", "Project Expo")

print("\nEvents Tuple:", events)
print("First Event:", events[0])
print("Last Event:", events[-1])
print("Total Number of Events:", len(events))

# 3. Departments Set Operations
departments = ["CSE", "IT", "ECE", "CSE", "EEE", "IT"]

print("\nDepartments List with Duplicates:", departments)

# Convert list to set (removes duplicates)
dept_set = set(departments)
print("Departments Set:", dept_set)

# Add department
dept_set.add("MECH")
print("After Adding MECH:", dept_set)

# Remove department
dept_set.remove("EEE")
print("After Removing EEE:", dept_set)

# 4. Winners List using Tuples and Convert to Frozenset
winners = [
    (1, "Arun"),
    (2, "Priya"),
    (3, "Rahul")
]

print("\nWinners List:", winners)

# Convert to frozenset
winner_frozenset = frozenset(winners)

print("Winners Frozenset:")
print(winner_frozenset)
```
