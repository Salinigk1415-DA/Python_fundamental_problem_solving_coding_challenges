```python
# 1. Read Two Numbers and Perform All Arithmetic Operations

# Read two numbers
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

# Arithmetic operations
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

```python
# 2. Check the Data Type of the Value Entered

value = input("Enter a value: ")

# Check data type
try:
    val = int(value)
    print("Data type is Integer")
except:
    try:
        val = float(value)
        print("Data type is Float")
    except:
        print("Data type is String")
```

```python
# 3. Find the Length of a String and Count the Number of Words

text = input("Enter a string: ")

# Length of string
print("Length of string:", len(text))

# Word count
words = text.split()
print("Number of words:", len(words))
```

```python
# 4. Count Words in a Sentence

sentence = input("Enter a sentence: ")

# Count words
word_count = len(sentence.split())

print("Number of words:", word_count)
```

```python
# 5. Reverse Each Word in a Sentence

sentence = input("Enter a sentence: ")

# Reverse each word
words = sentence.split()

reversed_words = [word[::-1] for word in words]

result = " ".join(reversed_words)

print("Reversed words sentence:")
print(result)
```

```python
# 6. Check Whether a Character is a Vowel or Consonant

ch = input("Enter a character: ").lower()

if ch in ['a', 'e', 'i', 'o', 'u']:
    print("It is a vowel")
else:
    print("It is a consonant")
```

```python
# 7. Check if a String is a Palindrome

text = input("Enter a string: ")

# Check palindrome
if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```
