Write a function that accepts a list of employee sales.

If sales are greater than 50,000, bonus is 10% of sales

If sales are between 30,000 and 50,000, bonus is 5%

Otherwise, no bonus

```python
# 1. Function to find the largest of three numbers
def largest(a, b, c):
    return max(a, b, c)

print("Largest Number:", largest(10, 25, 15))


# 2. Function to calculate factorial using recursion
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print("Factorial of 5:", factorial(5))


# 3. Function to check whether a string is a palindrome
def is_palindrome(text):
    text = text.lower()
    return text == text[::-1]

print("Is 'madam' a palindrome?", is_palindrome("madam"))


# 4. Function that accepts a list and returns the sum of all elements
def list_sum(numbers):
    return sum(numbers)

print("Sum of List:", list_sum([10, 20, 30, 40]))


# 5. Function using *args to find the average of multiple numbers
def average(*args):
    return sum(args) / len(args)

print("Average:", average(10, 20, 30, 40, 50))


# 6. Count students who scored above 75
def count_above_75(marks):
    count = 0
    for mark in marks:
        if mark > 75:
            count += 1
    return count

student_marks = [65, 80, 90, 70, 85]
print("Students Scored Above 75:", count_above_75(student_marks))


# 7. Calculate total sales greater than a given threshold
def total_sales_above_threshold(sales, threshold):
    total = 0
    for sale in sales:
        if sale > threshold:
            total += sale
    return total

sales = [10000, 25000, 40000, 15000, 50000]
print("Total Sales Above 20000:", total_sales_above_threshold(sales, 20000))


# 8. Count the number of vowels in a string
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

print("Number of Vowels:", count_vowels("Python Programming"))


# 9. Return only even numbers in a new list
def even_numbers(numbers):
    evens = []
    for num in numbers:
        if num % 2 == 0:
            evens.append(num)
    return evens

print("Even Numbers:", even_numbers([1, 2, 3, 4, 5, 6, 7, 8]))


# 10. Calculate Monthly Bonus for Employees
def calculate_bonus(employee_sales):
    bonuses = []

    for sales in employee_sales:
        if sales > 50000:
            bonus = sales * 0.10
        elif 30000 <= sales <= 50000:
            bonus = sales * 0.05
        else:
            bonus = 0

        bonuses.append(bonus)

    return bonuses

employee_sales = [25000, 35000, 60000, 45000, 70000]

print("Employee Sales:", employee_sales)
print("Bonuses:", calculate_bonus(employee_sales))
```
